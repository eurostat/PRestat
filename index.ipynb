**4 ways to download data from Eurostat database** 

```r
library("restatapi")
restatapi::get_eurostat_data("agr_r_milkpr",
                             filters=c("BE$","Hungary"),
                             date_filter=c(2015,"2006-01",2008:2010,"2014-01-01"),
                             keep_flags=T)
```

```r
library("eurostat")
eurostat::get_eurostat("agr_r_milkpr",
                       filters=list(geo=c("BE","HU"),time=c(2015,"2006-01",2008:2010,"2014-01-01")),
                       keepFlags=T)
```

```r
library("rdbnomics")
rdbnomics::rdb('Eurostat','agr_r_milkpr',dimensions=list(geo=c("BE","HU")))
```

```r
require("TSsdmx")
eurostat<-TSconnect("sdmx",dbname="EUROSTAT")
TSget("agr_r_milkpr....BE+HU",eurostat,start=2008,end=2010)
```
